#### Problem 1

```python
# create a dictionary of dataframes, one for each sheet in the Excel file (a dataframe is similar to an SPSS table)
# the keys for the dictionary are the names of the sheets; the dataframes can be called using the keys
# call the dataframe for the first problem; display the dataframe

df_dict = pd.read_excel("C:/Users/rmbrm/Documents/DS_WP/WPS/data-wrangling-input.xlsx", sheet_name=None)
input1 = df_dict.get('input1')
input1
```

```python
# group the data by form and create a list of mean scores by form

form = input1.groupby('form')
means = form.score.mean()
```

```python
# create a second dataframe listing the three values for the missing rows;
# round and format the calculated data as specified (i.e., no trailing zero)
# display the new dataframe

input1_md = pd.DataFrame({'group' : [3, 3],
                         'score': [(round(means['C']).astype('Int64')), (round(means['D']).astype('Int64'))],
                         'form': ["C", "D"]})
input1_md
```

```python
# Combine and sort the dataframes

input1.append(input1_md, ignore_index = True).sort_values(by=['group', 'form']) 
```

```python
# read in two dataframes

lookup = df_dict.get('input4b')
scores = df_dict.get('input4a')
```

```python
# using the lookup table, assign values for raw score conversion calculations
# display the values and score table

cs1x = (lookup.iloc[0][1])
cs1y = (lookup.iloc[0][2])
cs2x = (lookup.iloc[1][1])
cs2y = (lookup.iloc[1][2])

print (cs1x, cs1y, cs2x, cs2y)
scores
```

```python
# compute converted scores and fill values into score table

values = {'CS1': (cs1x * cs1y * scores['rawscore']),
          'CS2': (cs2x * cs2y * scores['rawscore'])}

scores.fillna(value=values)
```
